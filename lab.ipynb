```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('company_sales_data.csv')
```

```python
ax = plt.axes()
plt.xlabel("profit in doller")
plt.ylabel("month number")
plt.title("company profit month")
ax.scatter(df["month_number"],df["total_profit"],c='r', marker='o')
ax.scatter(df["month_number"],df["total_profit"],c='black', marker='.');
ax.plot(df["month_number"],df["total_profit"],c='r',linestyle='dashed')
plt.show()
```

```python
ax = plt.axes()
plt.xlabel("profit in doller")
plt.ylabel("month number")
plt.title("company profit month")
ax.plot(df["month_number"],df["total_profit"],c='r',linestyle='dashed', marker='o',markerfacecolor='black', markersize=6)
plt.show()
```

```python
df2=df.drop(['month_number','total_units','total_profit'],axis=1)
```

```python
df[df2.columns].plot(marker='o',title="sales data",ylabel="sales unit in number",xlabel="month number")
```

```python
ax2 = plt.axes()
plt.ylabel("sales unit in number")
plt.xlabel("month number")
plt.title("sales data")
for x in df2.columns:
    ax2.plot(df["month_number"],df[x])
    ax2.scatter(df["month_number"],df[x], marker='o')
plt.show()
```

```python
df3=np.array([1,2,3,4,5])
```

```python
fog,ax= plt.subplots(2,2)
ax[0,0].plot(df3,df3,c="black",linestyle="dashed")
ax[0,0].scatter(df3,df3,marker="*",c="black")
ax[0,1].plot(df3,df3*2)
ax[0,1].scatter(df3,df3*2,marker="o")
ax[1,0].plot(df3,df3**3,'o-.',c='yellow' , marker='s', markerfacecolor='r', markersize=6)
ax[1,1].plot(df3,df3**2,c="g",linestyle="dashed")
ax[1,1].scatter(df3,df3**2,c='g',marker="^")
```
