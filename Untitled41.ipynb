```python
import numpy as np
class pltt:
    def __init__(self,name,x,y,y2,n_x,n_y,n_y2):
        self.Name=name
        self.X=x
        self.Y=y
        self.Y2=y2
        self.N_x=n_x
        self.N_y=n_y
        self.N_y2=n_y2   
    def line(self,siz,color1,color2,lb1,lb2,lg):
        import matplotlib.pyplot as plt
        import math
        x_vals = self.X
        y_vals = self.Y
        y2_vals = self.Y2
        if(siz!=""):
            plt.rcParams["figure.figsize"] = siz
        plt.xlabel(self.N_x )
        plt.ylabel(self.N_y )
        plt.title(self.Name )
        plt.plot(x_vals, y_vals, color1 , label = lb1 )
        if(self.Y2!=""):
            plt.plot(x_vals, y2_vals, color2 , label = lb2 )
        if(lg==1):
            plt.legend(loc='upper center' )
    def scat(self,color1):
        import matplotlib.pyplot as plt
        import math
        x_vals = self.X
        y_vals = self.Y
        plt.xlabel(self.N_x)
        plt.ylabel(self.N_y)
        plt.title(self.Name )
        plt.scatter(x_vals, y_vals,c=color1 )
    def bar(self,color1):
        import matplotlib.pyplot as plt
        import math
        x_vals = self.X
        y_vals = self.Y
        plt.xlabel(self.N_x )
        plt.ylabel(self.N_y )
        plt.title(self.Name )
        plt.bar(x_vals, y_vals)
class hist:
    def __init__(self,name,data):
        import matplotlib.pyplot as plt
        import math
        plt.title(name )
        plt.hist(data )
class pi:
    def __init__(self,lab,val,exp):
        import matplotlib.pyplot as plt
        import math
        labels = lab
        values = val
        explode = exp
        plt.pie(values, explode=explode, labels=labels,autopct='%1.1f%%' , shadow=True)
        plt.show()
class dist:
    def __init__(self,data,siz=[10,8]):
        import matplotlib.pyplot as plt
        import seaborn as sns
        import math
        plt.rcParams["figure.figsize" ] = siz
        sns.distplot(df['petal_length' ])
class join:
    def __init__(self,name,X,Y,num=0):
        import seaborn as sns
        import math
        Data=sns.load_dataset(name)
        k=['scatter','reg','kde','hist','hex','resid' ]
        sns.jointplot(x=X , y=Y , data=Data , kind = k[num] )
        print("['scatter','reg','kde','hist','hex','resid' ]")
class pir:
    def __init__(self,name):
        import seaborn as sns
        import math
        Data=sns.load_dataset(name)
        sns.pairplot(data=Data)
class bar:
    def __init__(self,name,X,Y,siz=[8,6],style=1):
        import seaborn as sns
        import math
        Data=sns.load_dataset(name)
        plt.rcParams["figure.figsize" ] = siz
        st=["darkgrid", "whitegrid", "dark", "white", "ticks"]
        sns.set_style(st[style] )
        sns.barplot(x=X , y=Y , data=Data)
class coun:
    def __init__(self,name,X,Y,siz=[8,6],style=1):
        import seaborn as sns
        import math
        Data=sns.load_dataset(name)
        plt.rcParams["figure.figsize" ] = siz
        st=["darkgrid", "whitegrid", "dark", "white", "ticks"]
        sns.set_style(st[style] )
        sns.countplot(x=X  , data=Data)
snsc=["hist","pi->pieدائرة","dist-distplot","join->plot","pir->pairplot","bar->plot","coun->plot"]
plttm=["line","scat","bar"]
```
